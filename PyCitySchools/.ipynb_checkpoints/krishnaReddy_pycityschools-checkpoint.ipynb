```python
# Krishna Reddy 07/10/22
#pandas challenge ga tech homework
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
school_data_two = pd.read_csv(school_data_to_load)
student_data_two = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete_two = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#count number of schools, students. Sum budgets. average test scores
number_of_schools = school_data_complete["school_name"].nunique()
number_of_students=school_data_complete["Student ID"].nunique()
total_budget=school_data["budget"].sum()
avg_math_scores=student_data["math_score"].mean()
avg_reading_scores=student_data["reading_score"].mean()


#create filter for passing math scores, then count number of students
passing_math_scores=student_data["math_score"]>=70
amount_passing_math=passing_math_scores.sum()

#create filter for passing reading scores, then count number of students
passing_reading_scores=student_data["reading_score"]>=70
amount_passing_reading=passing_reading_scores.sum()


#set up conditional to identify students passing math and reading.  
#Apply boolean values based on conditional test. Add the booleans to the data
conditions=[(student_data["math_score"]>=70),
            (student_data["reading_score"]>=70)]

choices=[True,False]
student_data['Passing Math and Reading']=np.select(conditions,choices,default=True)

passing_both=(student_data['Passing Math and Reading']==True).count()


print("DISTRICT SUMMARY")
print("-----------------------------------------------------------------------")
print("Number of schools in district:      " + str(number_of_schools) + " schools.")
print("Number of students in district:     " + str(number_of_students) + " students.")
print("School district total budget:       $" + str(total_budget)+".")
print("Average Math Score:                 " + str(avg_math_scores)+".")
print("Average Reading Score:              " + str(avg_reading_scores)+".")
print("Students Passing Math:              " + str(amount_passing_math)+".")
print("Students Passing Reading:           " + str(amount_passing_reading)+".")
print("Students Passing Reading and Math:  " + str(passing_both)+".")
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#add boolean to combined dataset to create a value for students passing both math and reading
conditions=[(student_data["math_score"]>=70),
            (student_data["reading_score"]>=70)]

choices=[True,False]
school_data_complete['Passing Math and Reading']=np.select(conditions,choices,default=True)

#Set up variables to identify students passing individual subjects.  Add columns to the combined dataset.
passing_math= (school_data_complete["math_score"]>=70)
passing_reading= (school_data_complete["reading_score"]>=70)
school_data_complete["Passing Math"]=passing_math
school_data_complete["Passing Reading"]=passing_reading

#Calculate average values for school metrics
school_averages = school_data_complete.groupby(["school_name"]).mean()
averages_by_school=pd.DataFrame(school_averages).reset_index()


#Calculate budget per student
per_student_budget = school_averages["budget"]/school_averages["size"]
school_averages["Per Student Budget"]=per_student_budget



#Rename columns
Final_table=school_averages.rename(columns={"school_name":"School",
                                            "size":"School Size",
                                            "reading_score":"Avg Reading Score",
                                           "math_score":"Avg Math Score",
                                           "budget":"Total School Budget"})
#Remove unwanted columns
Final_table.drop(columns="Student ID", inplace=True)
Final_table.drop(columns="School ID", inplace=True)

#School type missing after averaging school values. Bring that column back in to the dataframe.
Final_table["Type"]=school_data["type"].values
Final_table
```

```python
budget_bins=[0,584,630,646,700]
budget_names=["<$584","$585-629","$630-644","$645-675"]
school_characteristics["Per Student Budget Ranges"]=pd.cut(school_characteristics["Per Student Budget"],
                                                           budget_bins,labels=budget_names,include_lowest=True)
budget_grouped=school_characteristics.groupby("Per Student Budget Ranges").mean()
budget_grouped
```
